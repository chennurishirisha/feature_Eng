OneHOtEncoding --variables with many categories

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/MercedesBenzTrain.csv', usecols =['X1','X2', 'X3', 'X4', 'X5', 'X6'])
```

```python
df.head()
```

```python
#lets have a look at how many labels each variables has

for col in df.columns:
  print(col, ':', len(df[col].unique()),'labels')
```

```python
#lets examine how many columns we will obtain after one hot encoding these variables
pd.get_dummies(df,drop_first =True).shape
```

we can see that from just 6 initial categorical variables we end up with 117 new variables.

```python
#lets find the top 10 most frequent categories for the variables X2
df.X2.value_counts().sort_values(ascending = False).head(20)
```

```python
#let's make a list with the most frequent categories of the variable
top_10 = [x for x in df.X2.value_counts().sort_values(ascending = False).head(10).index]
top_10
```

```python
for label in top_10:
  df[label] = np.where(df['X2']== label, 1, 0)

df[['X2'] + top_10].head(40)
```

```python
#get whole set of dummy variables for all categorical variables
def one_hot_top_x(df, variable,top_x_labels):
  for label in top_x_labels:
    df[variable + '_' + label] = np.where(df[variable] == label, 1, 0)
#read the data again
df = pd.read_csv('/content/MercedesBenzTrain.csv', usecols =['X1','X2', 'X3', 'X4', 'X5', 'X6'])
```

```python
one_hot_top_x(df,'X2', top_10)
df.head()
```

```python
# find the top 10 most frquent categories for X1
top_10 = [x for x in df.X1.value_counts().sort_values(ascending = False).head(10).index]

# now create the top most frequent dummy variables for x1

one_hot_top_x(df, 'X1',top_10)
df.head()
```
